# AMIML — arXiv Search AutomationAutomated retrieval of arXiv records matching the review's Boolean search strategy.Imports search and export functions from `amiml.arxiv`.

```python
import syssys.path.insert(0, "../src")from amiml.arxiv import search_arxiv, df_to_ris
```

## Define search queryThe master Boolean query combines five concept blocks (see §4.1.4 of the paper).

```python
QUERY = (    'all:"model-agnostic" OR all:"black-box" OR all:"surrogate"'    ' AND '    'all:"post-hoc" OR all:"post hoc"'    ' AND '    'all:"machine learning" OR all:"deep learning"'    ' AND '    'all:"interpretab" OR all:"explainab" OR all:"XAI"')print("Query:", QUERY[:120], "...")
```

## Execute search

```python
df = search_arxiv(QUERY, max_results=500)print(f"Retrieved {len(df)} records from arXiv")df[["title", "authors", "published"]].head()
```

## Export to RIS (for Rayyan import)

```python
n = df_to_ris(df, "../ris/identification/arxiv_export.ris")print(f"Exported {n} records to RIS")
```
